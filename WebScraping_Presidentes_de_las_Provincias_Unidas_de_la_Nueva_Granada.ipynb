```python
import urllib.request as urllib2
from bs4 import BeautifulSoup

import pandas as pd
```

```python
wikipage="https://es.wikipedia.org/wiki/Anexo:Presidentes_de_Colombia"
```

```python

page = urllib2.urlopen(wikipage)
```

```python

soup = BeautifulSoup(page, "lxml")
```

```python
#fina all the <table> tags
all_tables=soup.find_all('table')
all_tables
```

```python
#get the <table> tag that contains the data we want to scrap
right_table=soup.find('table', class_='wikitable')
right_table
```

```python
print (right_table.prettify())
```

```python
#To check the length of the table
for row in right_table.findAll("tr"):
    cells = row.findAll('td')
print(len(cells))
```

```python

#Generate lists
B=[]
D=[]
E=[]
F=[]
G=[]


#
for row in right_table.findAll("tr"):
    cells = row.findAll('td')
    if len(cells)==7: 
        b1=' '.join(cells[1].find_all(text=True))
        B.append(b1.replace('\n',""))
        d1=' '.join(cells[3].find_all(text=True))
        d1 = d1.replace("(e) ","") 
        D.append(d1.replace('\n',"")) 
        E.append(cells[4].find(text=True)) 
        f1=' '.join(cells[5].find_all(text=True))
        F.append(f1.replace('\n',""))
        g1=' '.join(cells[6].find_all(text=True))
        G.append(g1.replace('\n',""))
```

```python
print(len(G))
G
```

```python

df=pd.DataFrame(B, columns=['Periodo']) 
df['Presidente'] = D
df['Ciudad'] = E
df['Alineación (Partido Politico)'] = F
df['Comentarios'] = G

df
```

```python

df.to_csv('presidentes_nuevagranada.csv', mode='a', index=False)
```
