```python
import urllib.request as urllib2
from bs4 import BeautifulSoup

import pandas as pd
```

```python
wikipage="https://es.wikipedia.org/wiki/Anexo:Presidentes_de_Colombia"
```

```python

page = urllib2.urlopen(wikipage)
```

```python

soup = BeautifulSoup(page, "lxml")
```

```python
soup.title
```

```python
right_table=soup.find_all('table', class_='wikitable')
right_table[5]
```

```python
print (right_table[5].prettify())
```

```python
#To check the length of the table
for row in right_table[5].findAll("tr"):
    cells = row.findAll('td')
print(len(cells))
```

```python

#Generate lists
A=[]
B=[]
C=[]
D=[]
E=[]
F=[]

#
for row in right_table[5].findAll("tr"):
    cells = row.findAll('td')
    if len(cells)==6:
        b1=' '.join(cells[1].find_all(text=True))
        B.append(b1.replace('\n',""))
        d1=' '.join(cells[3].find_all(text=True))
        d1 = d1.replace("(e)","")
        D.append(d1.replace('\n',""))
        e1=' '.join(cells[4].find_all(text=True))
        E.append(e1.replace('\n',""))
        f1=' '.join(cells[5].find_all(text=True))
        F.append(f1.replace('\n',""))
```

```python
print(len(B))
B
```

```python

df=pd.DataFrame(B, columns=['Periodo']) 
df['Presidente'] = D
df['Partido'] = E
df['Comentarios'] = F



df
```

```python

df.to_csv('presidentes_republica.csv', mode='a', index=False)
```
